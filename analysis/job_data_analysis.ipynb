```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Data cleaning if necessary
file = "jobstreet_dataset.xlsx"
df = pd.read_excel(file)
print("Data imported successfully")

# Checking of null values
df.isnull().sum()
```

```python
sql_count = df['sql'].value_counts()['Yes']
excel_count = df['excel'].value_counts()['Yes']
tableau_count = df['tableau'].value_counts()['Yes']
power_bi_count = df['power_bi'].value_counts()['Yes']
python_count = df['python'].value_counts()['Yes']
r_count = df['r'].value_counts()['Yes']
```

```python
df_tech = pd.DataFrame(
   dict(
       tech = ['SQL', 'Excel', 'Tableau', 'Power BI', 'Python', 'R'],
       count = [sql_count, excel_count, tableau_count, power_bi_count, python_count, r_count]
   )
)

df_tech_sorted = df_tech.sort_values(by='count', ascending=False)
plt.bar('tech', 'count', data=df_tech_sorted, color='blue')
plt.title('Number of Jobs Requiring Specific Data Analytics Skills')  
plt.xlabel('Technology')  
plt.ylabel('Count')  
plt.show()
```
